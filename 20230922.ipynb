```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import Adam
```

```python
# MNIST 구현  (데이터 준비)

# Raw Data Loading
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ICT Innovation Square/data/mnist_train.csv')

# 결측치와 이상치 없음

# x_data와 t_data분리
x_data = df.drop('label', axis = 1, inplace = False).values # 2차원 ndarray
t_data = df['label'].values.reshape(-1, 1)  # 2차원 ndarray

# 정구화
scaler_x = MinMaxScaler()
scaler_x.fit(x_data)

x_data_norm = scaler_x.transform(x_data)

# train data test data 분리
x_data_train_norm, x_data_test_norm, t_data_train, t_data_test = train_test_split(x_data_norm, t_data, test_size = 0.3)
```

```python
# Logistic Regression (Multinomial Classification)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import Adam

logistic_model = Sequential()
logistic_model.add(Flatten(input_shape = (784, )))
logistic_model.add(Dense(units = 10, activation = 'softmax'))

logistic_model.compile(optimizer = Adam(learning_rate = 1e-2), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
logistic_model.summary()

# Total params: 7850 (30.66 KB)
# Trainable params: 7850 (30.66 KB)
# Non-trainable params: 0 (0.00 Byte)

logistic_model.fit(x_data_train_norm, t_data_train, epochs = 50, verbose = 1, batch_size = 100, validation_split = 0.3)

logistic_model_result = logistic_model.evaluate(x_data_test_norm, t_data_test)
print(logistic_model_result) # [0.4228643476963043, 0.9041270017623901]
```

```python
# DNN(Deep Nueral Network)구현
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import Adam

dnn_model = Sequential()
# input layer
dnn_model.add(Flatten(input_shape = (784, )))

# hidden layer
dnn_model.add(Dense(units = 64, activation = 'relu'))
dnn_model.add(Dense(units = 128, activation = 'relu'))
dnn_model.add(Dense(units = 256, activation = 'relu'))

# output layer
dnn_model.add(Dense(units = 10, activation = 'softmax'))

dnn_model.compile(optimizer = Adam(learning_rate = 1e-2), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
dnn_model.summary()

# Total params: 94154 (367.79 KB)
# Trainable params: 94154 (367.79 KB)
# Non-trainable params: 0 (0.00 Byte)

dnn_model.fit(x_data_train_norm, t_data_train, epochs = 50, verbose = 1, batch_size = 100, validation_split = 0.3)

dnn_model_result = dnn_model.evaluate(x_data_test_norm, t_data_test)
print(dnn_model_result) # [0.39320477843284607, 0.9525396823883057]
```

```python
# CNN(Convolution Nueral Network)구현
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout
from tensorflow.keras.optimizers import Adam

cnn_model = Sequential()

# 첫번째 convolution layer 겸 input layer
cnn_model.add(Conv2D(filters = 32, kernel_size = (3, 3), activation = 'relu', input_shape = (28, 28, 1)))
# filters = 32 - 필터의 갯수, kernel_size = (3, 3) - 커널 사이즈

cnn_model.add(MaxPooling2D(pool_size = (2, 2)))

cnn_model.add(Conv2D(filters = 64, kernel_size = (3, 3), activation = 'relu'))

cnn_model.add(MaxPooling2D(pool_size = (2, 2)))

cnn_model.add(Conv2D(filters = 64, kernel_size = (3, 3), activation = 'relu'))

cnn_model.add(Flatten())

# overfitting을 줄이기 위한 dropout layer 추가
cnn_model.add(Dropout(rate = 0.5))

cnn_model.add(Dense(units = 256, activation = 'relu'))

cnn_model.add(Dense(units = 10, activation = 'softmax'))

cnn_model.summary()

# Total params: 206026 (804.79 KB)
# Trainable params: 206026 (804.79 KB)
# Non-trainable params: 0 (0.00 Byte)

cnn_model.compile(optimizer = Adam(learning_rate = 1e-3), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])

cnn_model.fit(x_data_train_norm.reshape(-1, 28, 28, 1), t_data_train, epochs = 50, verbose = 1, batch_size = 100, validation_split = 0.3)

cnn_model_result =cnn_model.evaluate(x_data_test_norm.reshape(-1, 28, 28, 1), t_data_test)
print(cnn_model_result)  # [0.046799980103969574, 0.9907143115997314]
```
