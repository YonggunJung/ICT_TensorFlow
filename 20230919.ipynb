```python
# Multinomial Classification 구현

# 필요한 module import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import linear_model
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import Adam
```

```python
# Raw data Loading
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ICT Innovation Square/data/bmi.csv', skiprows = 3) # skiprows = 3 : 행을 스킵한다 3 세개를
display(df)
```

```python
# Data Preprocessing
# 결측치 확인
df.isnull().sum()
```

```python
# 데이터 분리
# train test split
x_data_train, x_data_test, t_data_train, t_data_test = train_test_split(df[['height', 'weight']], df['label'], test_size = 0.3)

# tensorflow 구현을 위한 정규화
scaler_x = MinMaxScaler()
scaler_x.fit(x_data_train)

x_data_train_norm = scaler_x.transform(x_data_train)
x_data_test_norm = scaler_x.transform(x_data_test)
```

```python
# sklearn 구현
sklearn_model = linear_model.LogisticRegression()

sklearn_model.fit(x_data_train, t_data_train)

# 모델 성능 평가
predict_value = sklearn_model.predict(x_data_test)

# accuracy score
from sklearn.metrics import accuracy_score

sklearn_result = accuracy_score(predict_value, t_data_test)
print(sklearn_result)
```

```python
# Tensorflow 구현

keras_model = Sequential()
keras_model.add(Flatten(input_shape = (2, )))
keras_model.add(Dense(units = 3, activation = 'softmax'))
keras_model.compile(optimizer = Adam(learning_rate = 1e-2), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy']) # sparse는 keras가 알아서 원핫 인코딩 해줌

keras_model.fit(x_data_train_norm, t_data_train, epochs = 200, verbose = 1, validation_split = 0.3)
```

```python
# 평가
result = keras_model.evaluate(x_data_test_norm, t_data_test)
print(result)
```

```python
# 예측
height = 187
weight = 81


my_state = np.array([[height, weight]])
predict_result = keras_model.predict(scaler_x.transform(my_state))

print(predict_result)
```
